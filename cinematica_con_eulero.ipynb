```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_array = np.linspace(0,100,10000)
```

```python

x = [10000]
v = [0]

for i in range(1,len(time_array)):
    
    a = -9.81

    dt = time_array[i] - time_array[i-1]
    
    new_v = v[i-1] + a * dt

    new_x = x[i-1] + new_v * dt
    
    v.append(new_v)
    x.append(new_x)

```

```python
plt.plot(time_array,x)
plt.ylim(0,11000)
```

# Moto di caduta libera con attrito dinamico

$$F = -\gamma\ v$$

```python
x = [10000]
v = [0]
a = [9.81]

gamma = 0.1

for i in range(1,len(time_array)):
    
    a = -9.81 - gamma * v[i-1]

    dt = time_array[i] - time_array[i-1]
    
    new_v = v[i-1] + a * dt

    new_x = x[i-1] + new_v * dt
    
    v.append(new_v)
    x.append(new_x)
```

```python
plt.plot(time_array,x)
plt.ylim(0,11000)
```

# Moto armonico

```python
x = [1]
v = [0]
#a = []

k = 1
gamma = 0.5

A_F = 1
w_F = 15

for i in range(1,len(time_array)):

    a = -k * x[i-1] - gamma * v[i-1] + A_F * np.cos(w_F*time_array[i-1])

    dt = time_array[i] - time_array[i-1]
    
    new_v = v[i-1] + a * dt

    new_x = x[i-1] + new_v * dt
    
    v.append(new_v)
    x.append(new_x)
```

```python
plt.plot(time_array,x)
```
