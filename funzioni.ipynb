```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch
```

```python
x = np.linspace(-5,5)
def f(x):
    return x**3
q = f(x) + 2
r = f(x) - 2




```

```python
fig, ax = plt.subplots(1, 3, figsize=(20,10))

ax[0].plot(x, f(x), label="$f(x)=x^3$",color='red')
ax[0].plot(x, f(x)+2, label="$f(x)+2$",color='blue')
ax[0].plot(x, f(x)-2, label="$f(x)-2$",color='green')

ax[1].plot(x, f(x), label="$f(x)=x^3$",color='red')
ax[1].plot(x, f(x-2), label="$f(x-2)$",color='purple')
ax[1].plot(x, f(x+2), label="$f(x+2)$",color='orange')

ax[2].plot(x, f(x), label="$f(x)=x^3$",color='red')
ax[2].plot(x, -f(x), label="$-f(x)$",color='navy')
ax[2].plot(x, f(x+2)+1, label="$f(x+2)+1=x^3+6x^2+12x+9$",color='lime')

# Assi uguali


# Etichette e titolo

#plt.title("Traslazioni della funzione f(x)=x")

import matplotlib.ticker as ticker
for i in range(3):
    ax[i].axhline(y=0,color='black',lw=1)
    ax[i].axvline(x=0,color='black',lw=1)
    ax[i].set_aspect("equal", adjustable="box")
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("y")
    ax[i].set_ylim(-10,10)
    ax[i].grid(True)
    ax[i].xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax[i].yaxis.set_major_locator(ticker.MultipleLocator(1))
    ax[i].legend()



arrow = FancyArrowPatch(
    posA=(0,0),
    posB=(0,2),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='blue'
)
ax[0].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[0].text(-0.2, 1, "+2", ha='right', va='center',color='blue')

arrow = FancyArrowPatch(
    posA=(0,0),
    posB=(0,-2),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='green'
)
ax[0].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[0].text(-0.2, -1, "-2", ha='right', va='center',color='green')


######################################################################

arrow = FancyArrowPatch(
    posA=(0,0),
    posB=(2,0),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='purple'
)
ax[1].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[1].text(1, 0.2, "2", ha='right', va='center',color='purple')

arrow = FancyArrowPatch(
    posA=(0,0),
    posB=(-2,0),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='orange'
)
ax[1].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[1].text(-1, 0.2, "2", ha='right', va='center',color='orange')



arrow = FancyArrowPatch(
    posA=(0,0),
    posB=(-2,0),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='lime'
)
ax[2].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[2].text(-1, 0.2, "2", ha='right', va='center',color='lime')

arrow = FancyArrowPatch(
    posA=(-2,0),
    posB=(-2,1),
    arrowstyle="->",
    mutation_scale=15,
    linewidth=3,
    color='lime'
)
ax[2].add_patch(arrow)

# Annotation *to the left* of the arrow
ax[2].text(-2.2, 0.5, "+1", ha='right', va='center',color='lime')





plt.show()
```
